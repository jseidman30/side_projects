```python
import pandas as pd
```

```python
from pathlib import Path
```

```python
filepath = "/Users/jeremyseidman/Desktop/Trilogy/cu-nyc-virt-fin-pt-03-2021-u-c/02-Homework/02-Python/Instructions/PyBank/Resources/budget_data.csv"

dataframe = pd.read_csv(filepath,index_col='Date', infer_datetime_format=True, parse_dates=True)
```

```python
dataframe["+/- change"] = (dataframe["Profit/Losses"] - dataframe["Profit/Losses"].shift(1))
```

```python
dataframe.head
```

```python
dataframe.head(87)
```

```python
totalmonths = len(dataframe)
totalmonths
```

```python
nettotal = sum(dataframe["Profit/Losses"])
nettotal
```

```python
maxprofit = 0
maxloss = 0
row = 0

maxprofit = dataframe["+/- change"].max()
minprofit = dataframe["+/- change"].min()

print(maxprofit)
print(minprofit)

dataframe = dataframe.dropna()

avgchange = sum(dataframe["+/- change"])/(totalmonths)
print(avgchange)

monthmost = dataframe.loc[dataframe["+/- change"]==maxprofit] 
print(monthmost)

monthleast = dataframe.loc[dataframe["+/- change"]==minprofit] 
print(monthleast)
```

```python
txtfile = Path("analysis.txt")
```

```python
with open (txtfile, "w") as txt:
    txt.write("Financial Analysis \n")
    txt.write("---------------------------- \n")
    txt.write(f"Total Months: {totalmonths} \n")
    txt.write(f"Total: ${nettotal:.2f} \n")
    txt.write(f"Average  Change: ${avgchange:.2f} \n")
    txt.write(f"Greatest Increase in Profits: \n {monthmost} \n")
    txt.write(f"Greatest Decrease in Profits: \n {monthleast} \n")
```
